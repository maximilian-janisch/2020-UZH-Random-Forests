# Coding part

## Reading in the data

```python
# Preliminary code, libraries we want to load
import os

import numpy as np
import pandas as pd

from sklearn.feature_selection import SelectKBest, f_classif

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Preliminary code, global variables
DATA_FILE = os.path.join("data", "dataset20.csv")
```

We will use the library _pandas_ for our data management.

```python
data = pd.read_csv(DATA_FILE, delimiter=";")
x = data.drop(columns="class") # put all information besides the class column in x
y = data["class"] # store the class column in y
```

As a brief sanity check: We can have a look at the dataset by printing the first five lines of data.

```python
print("First five lines of covariates:", x.head())
print("--------------------------------------------")
print("First five labels:", y.head())
```

Here is the shape of our data:

```python
print("Shape of the covariates:", x.shape)  # Sanity check, to see what our data looks like.
print("Shape of the labels:", y.shape)
```

# Data preprocessing

#### Feature selection

We will perform an f-test for the significance of each feature on the class variable. Also, we will produce a scatterplot for each feature where the class variable is used for the hue of the plot markers.

```python
## The f-test
filter = SelectKBest(score_func=f_classif, k="all")
x_no_na = x.dropna()._get_numeric_data()
y_no_na = data.dropna()["class"]
fit = filter.fit(x_no_na, y_no_na)
print(pd.DataFrame({"Feature": x_no_na.columns, "Score": fit.scores_}))

## Scatter plot for visual aid to determine whether a feature is relevant or not
fig, axs = plt.subplots(nrows=7, ncols=2, figsize=(12, 28))
axs = np.array(axs).flatten()
for k, col in enumerate(x.columns):
    sns.scatterplot(np.arange(len(x[col])), x[col], hue=y, ax=axs[k])
plt.tight_layout()
```

Evidently, not all features are relevant and we might as well _'forget'_ about them. We recall our features and mark them as relevant respectively irrelevant:

		pub_trans: public transport index  (relevant)
		price: price to rent per km (relevant)
		temperature: average tempereture during summer (relevant)
		inhabitants: number of inhabitants (irrelevant)
		registered: number of registered users in thousand (irrelevant)
		country: country (irrelevant)
		id: internal dataset code (irrelevant)
